```python
# create custom function


from langchain_community.tools import tool


@tool
def get_addtion():
    """
    thus tool is used for addtion or summation of two number
    passing two argument a and b

    Args:
        a (_type_): _description_
        b (_type_): _description_
    """
    a = input("Enter value of a")
    b = input("enter value of b")
    return int(a)+int(b)
```

```python
get_addtion.invoke('add two number')
```
