```python
from youtube_transcript_api import YouTubeTranscriptApi,TranscriptsDisabled
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_core.prompts import PromptTemplate
```

```python
import os
from langchain_groq import ChatGroq
from langchain.schema import Document

os.environ['GROQ_API_KEY'] = os.getenv("GROQ_API_KEY")
model = ChatGroq(model="llama3-70b-8192")




from langchain_community.vectorstores import Chroma
import os

from langchain.embeddings import HuggingFaceEmbeddings
huggingfacehub_api_token = os.getenv("HUGGINGFACEHUB_API_TOKEN")
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
video_id = "Gfr50f6ZBvo" # only the ID, not full URL
```

```python
try:
    transcript_list = YouTubeTranscriptApi.get_transcript(video_id=video_id,languages=['en'])
    print(transcript_list)

    transcript = " ".join(chunk['text'] for chunk in transcript_list)
    print(transcript)
except TranscriptsDisabled as e:
    print(e)
```

```python
transcript_list
```

```python
spliter = RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=20)


# convert into doc and create chunk
chunk = spliter.create_documents([transcript])
```

```python
vector_store = FAISS.from_documents(documents=chunk,embedding=embedding_model)
```

```python
vector_store.index_to_docstore_id
```

```python
vector_store.get_by_ids(['4f425dec-101d-47a8-9087-ccfa27b6a3bf'])
```

```python
retriever = vector_store.as_retriever(search_type='similarity',search_kwarg={"k":2})
```

```python
retriever.invoke("What is mean deepmind")
```

# Step 3 - Augmentation

```python
prompt = PromptTemplate(
    template=
    """
    You are smart and helpful assistance
    Answer only from provided transcripts contex.
    if the context is insufficient, Just says I don't Know.
    {context}
    Question:{question}
""")
```

```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableSequence,RunnableParallel,RunnablePassthrough,RunnableLambda

parser = StrOutputParser()
```

# Building Chain

```python
# first we need to extract the context


def formated_doc(retriever_text):
    context_text = "\n\n".join(doc.page_content for doc in retriever_text)
    return context_text
```

```python
paralle_chain = RunnableParallel(
    {"context":retriever | RunnableLambda(formated_doc),
     "question": RunnablePassthrough()
     }
)
```

```python
# test parallel chain

paralle_chain.invoke("what is deepmind")
```

```python
final_chain = paralle_chain | prompt | model |parser

# or

final_chain = RunnableSequence(paralle_chain,prompt,model,parser)
```

```python
final_chain.invoke("who is Demis")
```
